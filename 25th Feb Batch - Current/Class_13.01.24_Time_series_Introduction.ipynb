#### Time series analysis

-   Time series analysis is basically a regression problem

##### Time series Data Components -

-   Trend - Trend represents the general direction in which the data is moving , it can be upward , downward or flat
-   Seasonality - The repeating , fluctuations in the data that occurs on systematic regular intervals and follows a pattern.
-   Cycle - The repeating sequence of data are not tied to any specific calender intervals over here and also may not have any fixed duration.
-   Irregularity(Residuals) - random , unpredictable fluctuations in the data that are not defined by trend , seasonality or cycle

```python
import numpy as np 
import pandas as  pd 
import matplotlib.pyplot as plt 
```

```python
np.random.seed(42)
```

```python
n_points = 200 
date_rng = pd.date_range(start='2022-01-01',end='2022-07-19',freq='D')
```

```python
trend = 0.1*np.arange(n_points)
```

```python
seasonality = 10*np.sin(2*np.pi * np.arange(n_points)/30)
```

```python
cycle = 5*np.sin(2*np.pi * np.arange(n_points)/50)
```

```python
irregularity = np.random.normal(0,5,n_points)
```

```python
time_series_data = trend+seasonality+cycle+irregularity
```

```python
df = pd.DataFrame({
    'date':date_rng,
    'value':time_series_data
})
```

```python
df.head()
```

```python
plt.figure(figsize=(15,4))
plt.plot(df['date'],df['value'],label = 'Time series data')
plt.title('Synthetic time series data points')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,10))

#Trend
plt.subplot(4,1,1)
plt.plot(df['date'],trend,label='Trend')
plt.title('trend component')
plt.xlabel('date')
plt.ylabel('value')
plt.legend()

#seasonality
plt.subplot(4,1,2)
plt.plot(df['date'],seasonality,label='seasonality')
plt.title('seasonality component')
plt.xlabel('date')
plt.ylabel('value')
plt.legend()

#seasonality
plt.subplot(4,1,3)
plt.plot(df['date'],cycle,label='Cycle')
plt.title('cyclic component')
plt.xlabel('date')
plt.ylabel('value')
plt.legend()

#irregularity
plt.subplot(4,1,4)
plt.plot(df['date'],irregularity,label='residuals')
plt.title('residual component')
plt.xlabel('date')
plt.ylabel('value')
plt.legend()
```
