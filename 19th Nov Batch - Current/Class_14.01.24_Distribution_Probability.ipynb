##### Distribution 

-   Discrete Distribution
    *   Bernoulli Distribution - Imagine Flipping of a coin , This distribution helps us understand the chance of getting heads or tails  , it's like asking , will it rain tomorrow ? yes or no.. A random variable that can take one or two possible outcomes.
    *   Binomial Distribution - Imagine flipping the coin multiple times , and counting the chances of how many times you are getting heads .. If i flip the coin 4 times , how likely is it to get heads twice/thrice ? This distribution represents the number of successes in a fixed number of trials
    *   Poisson Distribution - Think about waiting for a bus in a bus station. This distribution helps us to know/understand how many buses can we expect to arrive in an hour ? - This describes the number of events that will occur within a fixed interval of time.
    *   Geometric Distribution - Suppose you are tossing a coin until you are getting heads. This distribution tells you how many times you need to toss the coin before getting heads. Determining the number of trials needed for a success to occur for the very first time.

```python
# Bernoulli Distribution

import numpy as np 
import matplotlib.pyplot as plt 

p = 0.3 # probability of success 
data_bernoulli = np.random.choice([0,1],size = 1000 , p=[1-p,p])

plt.hist(data_bernoulli,bins=[-0.5,0.5,1.5],edgecolor = 'black')
plt.title('Bernoulli Distribution')
plt.xlabel('Outcome')
plt.ylabel('frequency')
plt.show()
```

```python
# Binomial Distribution 

n = 5 #the number of trials 
p = 0.5 #probability of success in each trial
data_binomial = np.random.binomial(n,p,size = 1000)

plt.hist(data_binomial,bins = np.arange(0.5,n+1.5,1),edgecolor = 'black')
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

```python
# Poisson Distribution 

lambda_ = 3 
data_poission = np.random.poisson(lambda_,size = 1000)

plt.hist(data_poission,bins = np.arange(-0.5,15.5,1),edgecolor='black')
plt.title('poission distribution')
plt.xlabel('number of events')
plt.ylabel('frequency')
plt.show()
```

```python
mu = 0 
sigma = 1 
data_normal = np.random.normal(mu,sigma,size = 1000)

plt.hist(data_normal,bins = 30 ,edgecolor = 'black')
plt.title('normal distribution')
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

```python
# Geometric Distribution 

p = 0.3 
data_geometric = np.random.geometric    (p,size=1000)

plt.hist(data_geometric,bins = np.arange(0,15),edgecolor = 'black')
plt.title('geometric distribution')
plt.xlabel('Number of trials untill first success')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Uniform Distribution

low = 0
high = 1 
data_uniform = np.random.uniform(low,high,size=1000)

plt.hist(data_uniform , bins = 30 ,edgecolor = 'black')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# exponential distribution 
beta = 0.5
data_exponential = np.random.exponential(beta,size = 1000)

plt.hist(data_exponential , bins = 30 ,edgecolor = 'black')
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
k = 2 
theta = 1 
data_gamma = np.random.gamma(k,theta,size=1000)

plt.hist(data_gamma , bins = 30 ,edgecolor = 'black')
plt.title('Gamma Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
df = 3 
data_chi_square = np.random.chisquare(df,size = 1000)

plt.hist(data_chi_square , bins = 30 ,edgecolor = 'black')
plt.title('Chi Sqaue Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
