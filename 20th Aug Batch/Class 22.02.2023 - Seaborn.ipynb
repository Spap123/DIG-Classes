```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
#pointplot

sns.pointplot(y = titanic['age'] ,
 x = titanic['class'], hue = titanic['sex'] ,data = titanic).set_title('point plot age with class')
```

```python

sns.lineplot(
    y = titanic['fare'],
    x = titanic['embarked'],
    data = titanic).set_title('line plot for fare and embarked')
```

```python
import matplotlib.pyplot as plt
# FacetGrid
grid1 = sns.FacetGrid(titanic,col = 'class')
grid1.map(plt.hist,'age')
grid3 = sns.FacetGrid(titanic,col = 'class')
grid3.map(plt.scatter,'age','fare')
```
