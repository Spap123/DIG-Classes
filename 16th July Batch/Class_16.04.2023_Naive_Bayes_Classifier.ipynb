```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
cancer_df = pd.DataFrame(data = cancer.data , columns = (cancer.feature_names))
```

```python
cancer_df
```

```python
cancer_df['target'] = cancer.target
```

```python
cancer_df.head(25)
```

```python
x = cancer_df.drop('target',axis='columns')
y = cancer_df.target 
```

```python
x.shape
```

```python
y.shape
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
nb.fit(xtrain,ytrain)
```

```python
nb.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score 
from sklearn.metrics import accuracy_score , confusion_matrix , ConfusionMatrixDisplay
```

```python
# accuracy on training data
predict_train = nb.predict(xtrain)
accuracy_train = accuracy_score(ytrain,predict_train)
```

```python
predict_train
```

```python
ytrain
```

```python
print(accuracy_train)
```

```python
# accuracy on training data
predict_test = nb.predict(xtest)
accuracy_test = accuracy_score(ytest,predict_test)
```

```python
print(nb.score(xtest,ytest))
```

```python
print(accuracy_test)
```

```python
cv_score = cross_val_score(nb,x,y,cv = 5)
```

```python
print(cv_score)
```
