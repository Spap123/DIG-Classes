```python
# Task - Find out average salaries of  company's employees
# Data - 10 people salary
# executive - 25000
# sr executive - 30000
# operation executive - 18000
# executive - 26000
# sr executive - 32000
# operation executive - 19500
# executive - 25600
# sr executive - 33000
# operation executive - 19800
# CEO - 450000
```

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
salary = pd.DataFrame(data=[
    ['CEO',150000],
    ['executive',15000],
    ['sr executive',16000],
    ['operation executive',25000],
    ['sr executive',18000],
    ['director',180000],
    ['sr executive',15000],
    ['sr executive',17000]],columns = ['Designation','Salary'])
```

```python
salary.head()
```

```python
print(salary['Salary'].mean())
```

```python
sns.boxplot (salary['Salary'])
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,4))
ax.scatter(salary['Salary'],salary['Designation'])
```

```python
# IQR Method - interquartile range method
# All the data points are evaluated in the first.
# Sorting the data points in ascending order
# Q1 - 25th percentile , Q2 - 50th Percentile , Q3 - 75th perentile
```

```python
Q1 = np.percentile(salary['Salary'],25,interpolation='midpoint')
Q3 = np.percentile(salary['Salary'],75,interpolation='midpoint')
```

```python
Q1
```

```python
Q3
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
low_lim = Q1-(1.5*IQR)
upper_lim = Q3+(1.5*IQR)
```

```python
low_lim
```

```python
upper_lim
```

```python
outliers = []
for i in (salary['Salary']):
    if ((i>upper_lim) or (i<low_lim)):
        outliers.append(i)
```
