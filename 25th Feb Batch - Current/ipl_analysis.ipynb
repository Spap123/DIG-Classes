```python
# Create a new column indicating the match outcome: 1 for win, 0 for tie, and -1 for no result
ipl['match_outcome'] = ipl['result'].map({'normal': 1, 'tie': 0, 'no result': -1})

# Create a new column indicating if the team that won the toss also won the match
ipl['toss_win_match_win'] = ipl.apply(lambda row: 1 if row['toss_winner'] == row['winner'] else 0, axis=1)

# Calculate the correlation
correlation = ipl['toss_win_match_win'].corr(ipl['match_outcome'])


# Create a cross-tabulation between toss win and match outcome
cross_tab = pd.crosstab(ipl['toss_win_match_win'], ipl['match_outcome'])

# Create a clustered bar plot
ax = sns.heatmap(cross_tab, annot=True, fmt="d", cmap="coolwarm", cbar=True)
plt.title("Correlation Between Toss Win and Match Outcome")
plt.xlabel("Match Outcome")
plt.ylabel("Toss Win")
plt.show()
```

```python

# Create a new column indicating if the team that won the toss also won the match
ipl['toss_win_match_win'] = ipl.apply(lambda row: 1 if row['toss_winner'] == row['winner'] else 0, axis=1)

# Create a grouped bar plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Group the data by toss decision and calculate the count of wins and losses
grouped = ipl.groupby(['toss_decision', 'toss_win_match_win']).size().reset_index(name='count')

# Define custom colors for the legend
custom_palette = {0: "lightcoral", 1: "lightseagreen"}

# Create the bar plot
ax = sns.barplot(x='toss_decision', y='count', hue='toss_win_match_win', data=grouped, palette=custom_palette)

# Set labels and title
ax.set_xlabel("Toss Decision")
ax.set_ylabel("Count")
ax.set_title("Toss Decision vs. Match Outcome")

# Add labels for the bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2, height), ha='center', va='bottom')

# Customize the legend
legend_labels = ["Match Loss", "Match Win"]
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles, legend_labels, title="Match Win", loc="upper right")

# Show the plot
plt.show()
```

```python
# Group the data by 'player_of_match' and count the number of awards for each player
player_of_match_counts = ipl['player_of_match'].value_counts()

# Sort the players by the number of awards in descending order
sorted_players = player_of_match_counts.sort_values(ascending=False)

# Create a function to plot the data
def plot_top_players(num_players):
    plt.figure(figsize=(12, 6))
    plt.bar(sorted_players.index[:num_players], sorted_players.values[:num_players], color='skyblue')
    plt.xlabel('Player Names')
    plt.ylabel('Number of Player of the Match Awards')
    plt.title(f'Top {num_players} Players with Player of the Match Awards')
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.show()

# Create a slider widget for selecting the number of players to display
num_players_slider = widgets.IntSlider(value=10, min=1, max=len(sorted_players), step=1, description='Top Players:')

# Create an interactive plot using the widget value
widgets.interactive(plot_top_players, num_players=num_players_slider)
```

```python
winning_team_data = ipl[['city','winner']]
# Group the data by 'city' and 'winner' and calculate the count
city_winner_counts = winning_team_data.groupby(['city', 'winner']).size().reset_index(name='count')

# Pivot the table to have cities as rows and winners as columns
city_winner_counts_pivot = city_winner_counts.pivot_table(index='city', columns='winner', values='count', fill_value=0)

# Display the result
print(city_winner_counts_pivot)
```

```python
# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(city_winner_counts_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)
plt.title('Count of Winners in Each City')
plt.xlabel('Winner')
plt.ylabel('City')
plt.xticks(rotation=90)
plt.show()
```

```python

# Calculate the count of matches with and without DL applied
dl_counts = ipl['dl_applied'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(dl_counts, labels=['DL Not Applied', 'DL Applied'], autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('DL Applied vs. DL Not Applied Ratio')
plt.show()
```

```python

# Group the data by 'Season' and calculate the count of matches in each season
season_matches_count = ipl['Season'].value_counts().reset_index()
season_matches_count.columns = ['Season', 'Matches']

# Sort the seasons by year
season_matches_count = season_matches_count.sort_values(by='Season')

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Season', y='Matches', data=season_matches_count, palette="viridis")
plt.title('Number of Matches Played in Each Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```
