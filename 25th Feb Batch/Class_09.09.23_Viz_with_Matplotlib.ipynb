```python
fig,axes = plt.subplots(2,2)
data1 = [10,5,8,12]
data2 = [5,8,6,9]
data3 = [8,6,12,7]
data4 = [3,6,9,4]

axes[0,0].bar(['a','b','c','d'],data1)
axes[0,0].set_title('subplot1')
axes[0,1].bar(['a','b','c','d'],data2)
axes[0,1].set_title('subplot2')
axes[0,0].bar(['a','b','c','d'],data3)
axes[0,0].set_title('subplot3')
axes[0,1].bar(['a','b','c','d'],data4)
axes[0,1].set_title('subplot4')

plt.show()
```

```python
countries = ['A','B','C','D','E']
population_1930 = [10,20,30,40,50]
population_1940 = [10,60,30,80,50]
population_1950 = [10,20,40,40,50]
population_1960 = [10,20,90,40,40]
population_1970 = [10,30,50,90,70]
```

```python
plt.bar(countries,population_1930,color='Green')
plt.bar(countries,population_1940,color='yellow')
plt.bar(countries,population_1950,color='blue')
plt.bar(countries,population_1960,color='brown')
plt.bar(countries,population_1970,color='pink')
plt.xlabel('countries')
plt.ylabel('population over the years')
plt.title('stacked bar chart')
plt.show()
```

```python

x = [8,9,6,2,4]
y = [4,5,9,2,4]
a = [9,7,5,6,3]
b = [3,9,6,5,8]

plt.plot(x,y,color='k',linestyle='--',marker='.',label="plot xy")
plt.plot(a,b,label='plot ab')
plt.legend(labelcolor = 'black')
plt.title('multiline graph')
plt.show()
```

```python
x = [1,3,5,7,9]
y = [2,4,3,1,7]

x2 = [2,4,6,8,10]
y2 = [2,4,4,2,6]

plt.bar(x,y,label='first bars')
plt.bar(x2,y2,label= 'second bars')

plt.title('multiple bar charts')
plt.legend()
plt.show()
```
